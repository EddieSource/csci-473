```python
# Numerical/scientific computing packages.
import numpy as np
import scipy
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import sklearn

from sklearn.linear_model import LinearRegression, LogisticRegression

print('Finished successfully loading packages')
```

```python
# import datasets
data = np.genfromtxt('housingUnits.csv', delimiter=',')
print(data.shape)
```

```python
# question 2
# standardize column 2 and column 3
data_cleaned = data.copy()[1:,:];
print(data_cleaned.shape)

# divide # of roooms and bedrooms in a block and by # of household and population in a block
fig, axes = plt.subplots(2,2,figsize=(15,10))

features = ["total rooms per population", "total bedrooms per population", "total rooms per household", "total bedrooms per household", ]
for i in range(2): 
    for j in range(2): 
        data_cleaned[:,j + 1] = data[1:,j + 1]/data[1:,i + 3]
        X = data_cleaned[:,j + 1].reshape(-1,1)
        y = data_cleaned[:,7].reshape(-1,1)
        model = LinearRegression().fit(X, y) # train part
        rSqr = model.score(X, y)
        axes[i][j].plot(X, y,'o',markersize=.75)
        axes[i][j].set_xlabel(features[i * 2 + j]) 
        axes[i][j].set_ylabel('median house value')  
        axes[i][j].set_title('R^2 = {:.3f}'.format(rSqr))

# clean the data using standardized value
data_cleaned[:,1] = data[1:,1]/data[1:,3]
data_cleaned[:,2] = data[1:,2]/data[1:,3]
```

```python
# question 3
# use linear regression on each predictor to get outcome
# use matplotlib to visualize scatterplots
import matplotlib
import matplotlib.pyplot as plt

features = ["housing median age", "total rooms per population", "total bedrooms per population", "population", "household", "median income", "ocean proximity"]

fig, axes = plt.subplots(4,2,figsize=(13,25))
for i in range(7): 
    X = data_cleaned[:,i].reshape(-1,1)
    y = data_cleaned[:,7].reshape(-1,1)

    model = LinearRegression().fit(X, y) # train part
    rSqr = model.score(X, y)
    
    axes[i//2][i%2].plot(X, y,'o',markersize=.75)
    axes[i//2][i%2].set_xlabel(features[i]) 
    axes[i//2][i%2].set_ylabel('median house value')  
    axes[i//2][i%2].set_title('R^2 = {:.3f}'.format(rSqr))
    
    
```

```python
# question 4
# run a multiple regression model
X = data_cleaned[:,:7]
y = data_cleaned[:,7].reshape(-1,1)

model = LinearRegression().fit(X, y) # train part
rSqr = model.score(X,y)
yHat = model.predict(X)

# Visualize: actual vs. predicted income (from model)
plt.figure()
plt.plot(yHat,y,'o',markersize=.75) 
plt.xlabel('Predicted value') 
plt.ylabel('Actual value')  
plt.title('R^2 = {:.3f}'.format(rSqr))
```

```python
# Question 5
# run a linear regression between variables 2 and 3
X = data_cleaned[:,1]
y = data_cleaned[:,2]
corr = np.corrcoef(X, y)[0,1]

model = LinearRegression().fit(X.reshape(-1,1), y.reshape(-1,1))

fig, axes = plt.subplots(1,2,figsize = (10,5))

axes[0].plot(X,y,'o', markersize=.75)
axes[0].set_xlabel(features[1])
axes[0].set_ylabel(features[2])
axes[0].set_title('corr = {:.3f}'.format(corr))
          

# run a linear regression between variables 4 and 5
X = data_cleaned[:,3]
y = data_cleaned[:,4]
corr2 = np.corrcoef(X, y)[0,1]

model = LinearRegression().fit(X.reshape(-1,1), y.reshape(-1,1))

axes[1].plot(X,y,'o', markersize=.75)
axes[1].set_xlabel(features[3])
axes[1].set_ylabel(features[4])
axes[1].set_title('R^2 = {:.3f}'.format(corr2))
          
```
